```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from sklearn.naive_bayes import GaussianNB
%matplotlib inline
```

```python
data = pd.read_csv("iris5.csv",encoding='latin-1')
data.head()
```

```python
X = data.drop(['species'], axis = 1)
Y = data['species']
c=[]
for i in Y:
    if i==1:
        c.append('lightcoral')
    elif i==2:
        c.append('lightgreen')
    else:
        c.append('deepskyblue')
    
    
```

```python
plt.scatter(X['sepal_length'], X['sepal_width'],color=c)
```

```python
plt.scatter(X['petal_length'], X['petal_width'],color=c)
```

```python
clf = GaussianNB()
model = clf.fit(X, Y)
print(model.score(X, Y))
```

```python
p=clf.predict(X)
confusion_mat=np.column_stack((Y,p))
```
